1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

````python
A = np.array([1, 2, 3, 4, 5])
B = np.array([6, 7, 8, 9, 10])

AB_vert = np.vstack((A,B))
AB_horiz = np.hstack((A,B))

print(f"A and B stacked vertically is: \n {AB_vert}")
print(f"A and B stacked horizontally is: \n {AB_horiz}")
````

````python
A = np.array([1, 1, 3, 5, 7])
B = np.array([1, 4, 6, 7, 8])

C = np.intersect1d(A, B)

print(f"The array of common elements between A and B are {C}.")
````

4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0
```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

```

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

petal_length_filt = iris_2d[:,2][iris_2d[:,2] > 1.5]
iris_2d[:,2] = np.where(iris_2d[:,2] <= 1.5, iris_2d[:,2], np.nan)

sepal_length_filt = iris_2d[:,0][iris_2d[:,0] < 5.0]
iris_2d[:,0] = np.where(iris_2d[:,0] >= 5.0, iris_2d[:,0], np.nan)

print(f"The filtered petal lengths are: \n{petal_length_filt}")
print(f"The filtered sepal lengths are: \n{sepal_length_filt}")
print(f"The original table with filtered data removed is: \n{iris_2d}")

for el in iris_2d[:,2]:
    if el > 1.5 and el != np.nan:
        raise ValueError("Third column has misfiltered data with petal length <= 1.5")
for el in iris_2d[:,0]:
    if el < 5:
        raise ValueError("First column has misfiltered data with sepallength > 5.0")
````

2. Replace missing values in Min.Price and Max.Price columns with their respective mean.

```
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

````python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

# Replacing nan values with the mean of each column
min_price_series = df['Min.Price'][:].replace(np.nan, df['Min.Price'].mean()) 
max_price_series = df['Max.Price'][:].replace(np.nan, df['Max.Price'].mean())

# Forming a dataframe with the series
price_dict = {'Min.Price': min_price_series, 'Max.Price': max_price_series}
price_df = pd.DataFrame(data=price_dict)

# Overwriting original values with our substitutions
df[['Min.Price', 'Max.Price']] = price_df[['Min.Price', 'Max.Price']]

df
````

3. How to get the rows of a dataframe with row sum > 100?

```
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))

ans_df = df.where(df.sum(axis=1, numeric_only = True) > 100)
ans_df.dropna(how = 'all', inplace = True)
ans_df
````
