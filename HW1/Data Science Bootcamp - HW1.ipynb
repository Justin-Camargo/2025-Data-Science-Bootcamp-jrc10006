## Problem 1:

```python
print("Enter a word to get its vowel count")
word = input()

if(word.count(" ") > 0):
    raise Exception("Please, only enter one word.")

def get_vowel_count(word):
    vowel_count = 0;
    vowel_dictionary = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0, "y": 0}
    for character in word:
        match character:
            case 'a':
                vowel_count += 1
                vowel_dictionary.update({'a':vowel_dictionary.get('a')+1})
            case 'e':
                vowel_count += 1
                vowel_dictionary.update({'e':vowel_dictionary.get('e')+1})
            case 'i':
                vowel_count += 1
                vowel_dictionary.update({'i':vowel_dictionary.get('i')+1})
            case 'o':
                vowel_count += 1
                vowel_dictionary.update({'o':vowel_dictionary.get('o')+1})
            case 'u':
                vowel_count += 1
                vowel_dictionary.update({'u':vowel_dictionary.get('u')+1})
    if(vowel_count == 0):
        for character in word:
            if(character == 'y'):
                vowel_count += 1
                vowel_dictionary.update({'y':vowel_dictionary.get('y')+1})
    # return vowel_count
    return vowel_dictionary, vowel_count
vowel_dict, vowel_count = get_vowel_count(word)
print(f"The total number of vowels in {word} is {vowel_count}:")
for item in vowel_dict.items():
    print(f"\t The vowel count for {item[0]} is {item[1]}")
```
